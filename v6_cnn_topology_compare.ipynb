```python
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dropout, Dense
```

```python
from functions.data import load_samples_split

x_train, x_test, y_train, y_test = load_samples_split()
```

```python
from functions.visual import plot_loss, plot_accuracy, plot_confusion
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dropout(0.5))
model.add(Dense(512, activation='sigmoid'))
model.add(Dense(256, activation='sigmoid'))
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(2, activation='sigmoid'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, validation_split=0.2, epochs=150, verbose=False)
```

```python
model.evaluate(x_test, y_test, verbose=2)
```

```python
plot_accuracy(history)
```

```python
plot_loss(history)
```

```python
plot_confusion(model, x_test, y_test)
```
